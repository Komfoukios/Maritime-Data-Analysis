# Maritime Data Analysis
This project analyzes hypothetical maritime voyage data, including travel time and fuel consumption for different routes. Using Python (Pandas, Matplotlib), we visualize and extract insights from the dataset.


```python
# Εισάγουμε τις βιβλιοθήκες
import pandas as pd
import matplotlib.pyplot as plt

# Δημιουργούμε υποθετικό dataset
data = {
    'Departure_Port': ['Piraeus', 'Rotterdam', 'Hamburg', 'Marseille', 'Piraeus'],
    'Arrival_Port': ['Valencia', 'New York', 'Rotterdam', 'Algeciras', 'Hamburg'],
    'Travel_Time_hours': [72, 168, 48, 36, 96],
    'Fuel_Consumption_tons': [150, 600, 200, 120, 400]
}

# Μετατρέπουμε τα δεδομένα σε DataFrame
df = pd.DataFrame(data)

# Εμφανίζουμε το DataFrame
df
```

```python
# Υπολογισμός μέσης κατανάλωσης καυσίμου
mean_fuel = df['Fuel_Consumption_tons'].mean()
print(f"Average Fuel Consumption: {mean_fuel:.2f} tons")

# Εύρεση ταξιδιού με τη μέγιστη κατανάλωση
max_fuel_trip = df.loc[df['Fuel_Consumption_tons'].idxmax()]
print("Trip with Maximum Fuel Consumption:")
print(max_fuel_trip)
```

```python
# Δημιουργία γραφήματος
plt.figure(figsize=(10,5))
plt.bar(df['Departure_Port'] + ' - ' + df['Arrival_Port'], df['Fuel_Consumption_tons'], color='skyblue')
plt.title('Fuel Consumption per Route')
plt.xlabel('Route')
plt.ylabel('Fuel Consumption (tons)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
